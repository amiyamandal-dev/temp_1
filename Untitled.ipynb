````python
from mistralrs import Runner, Which, ChatCompletionRequest, VisionArchitecture
````

````python
MODEL_ID = "meta-llama/Llama-3.2-11B-Vision-Instruct"
````

````python
runner = Runner(
    which=Which.VisionPlain(
        model_id=MODEL_ID,
        arch=VisionArchitecture.VLlama,
    ),
    in_situ_quant="Q4K",
)
````

````python
%%time
res = runner.send_chat_completion_request(
    ChatCompletionRequest(
        model="llama-vision",
        messages=[
            {
                "role": "user",
                "content": [
                    {
                        "type": "image_url",
                        "image_url": {
                            "url": ""
                        },
                    },
                    {
                        "type": "text",
                        "text": "What is shown in this image? Write a detailed response analyzing the scene.",
                    },
                ],
            }
        ],
        max_tokens=256,
        presence_penalty=1.0,
        top_p=0.1,
        temperature=0.1,
    )
)
````

````python
import fitz  # PyMuPDF
import base64

class PDFPageIterator:
    def __init__(self, pdf_data):
        """
        Accepts a PDF file path or bytes data and creates an iterator over base64-encoded images and text of each page.

        :param pdf_data: Path to the PDF file or PDF data in bytes.
        """
        # Open the PDF document
        if isinstance(pdf_data, str):
            self.doc = fitz.open(pdf_data)
        elif isinstance(pdf_data, bytes):
            self.doc = fitz.open("pdf", pdf_data)
        else:
            raise ValueError("pdf_data must be a file path or bytes data.")

        self.page_count = len(self.doc)
        self.current_page = 0  # Initialize the page counter

    def __iter__(self):
        return self

    def __next__(self):
        if self.current_page >= self.page_count:
            raise StopIteration

        page = self.doc[self.current_page]
        self.current_page += 1

        # Render page to an image
        pix = page.get_pixmap()
        # Convert image to bytes in PNG format
        img_bytes = pix.tobytes("png")
        # Encode image to base64 string
        base64_str = base64.b64encode(img_bytes).decode('utf-8')

        # Extract text from the page
        text = page.get_text()

        return base64_str, text
````

````python
import copy
import tqdm
import json
````

````python
pdf_iterator = PDFPageIterator("aami-pds.pdf")
````

````python
all_rez = dict()
for page_count, (base64_str, text) in enumerate(pdf_iterator, start=1):
    prompt_val = f'''
Objective:
Convert the provided text into a structured HTML format, using the reference image to guide the HTML hierarchy and relationships.

---

Inputs:

1. Text:
   - The content to be converted into HTML.

2. Image:
   - A reference image illustrating the desired HTML structure, including hierarchy and grouping.

---

Instructions:

1. Analyze the Image:
   - Understand the desired HTML layout, focusing on hierarchy and grouping of elements.

2. Process the Text:
   - Identify all information from the text input to include in the HTML structure.

3. Create HTML Structure:
   - Construct a valid, well-formatted HTML document that accurately represents the text content.
   - Use the image solely as a guide for structuring HTML elements.
   - Do not include any information from the image that is not present in the text.

4. Add Page Number Attribute:
   - For each page represented in the image or text, add a `page_no` attribute to the `<body>` tag.
   - Example: `<body page_no="10">`

---

Rules:

- Content Inclusion:
  - Include only the information provided in the text input.
  - Do not infer or add any additional information from the image.

- Hierarchy and Grouping:
  - Reflect the hierarchy and grouping shown in the image within the HTML structure.

- Preserve Original Text:
  - Use the exact wording from the text input in the HTML content.

---

Output Format:

- Provide a valid, well-formatted HTML structure based solely on the text input, structured according to the image's guidance.
- Enclose content within `<body>` tags.
- For multiple pages, use separate `<body>` blocks with appropriate `page_no` attributes.

Example:

```html
<body page_no="1">
  <!-- Content for page 1 -->
</body>

<body page_no="2">
  <!-- Content for page 2 -->
</body>
```

---

Text Input:

{text}

---
'''
    res = runner.send_chat_completion_request(
        ChatCompletionRequest(
            model="llama-vision",
            messages=[
                {
                    "role": "user",
                    "content": [
                        {
                            "type": "image_url",
                            "image_url": {
                                "url": base64_str
                            },
                        },
                        {
                            "type": "text",
                            "text": prompt_val,
                        },
                    ],
                }
            ],
            max_tokens=512,
            presence_penalty=1.0,
            top_p=0.1,
            temperature=0.00,
        )
    )
    all_rez[page_count] = res.choices[0].message.content
    with open("aami-llama-13b-v.json", "w") as f:
        json.dump(all_rez, f)
````
